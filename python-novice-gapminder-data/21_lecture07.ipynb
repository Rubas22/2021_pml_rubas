Con make_blobs genera 1000 datos con semilla 33. Luego aplica el dataset a AgglomerativeClustering con les criterios de fusión: ward, average y complete (linkage) buscando 4 clusters. ¿Con cual de los criterios se obtiene un cluster muy pequeño?

Hint: grafica con: plt.scatter(X[:, 0], X[:, 1], c=prediction, s=60) donde la variable prediction es la prediccion del modelo (.fit_predicc)

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# generate synthetic two-dimensional data
X, y = make_blobs(n_samples=1000, random_state=33)
```

```python
agg = AgglomerativeClustering(linkage='ward', n_clusters=4)
assignment = agg.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=assignment, s=60)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
agg = AgglomerativeClustering(linkage='average', n_clusters=4)
assignment = agg.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=assignment, s=60)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

```python
agg = AgglomerativeClustering(linkage='complete', n_clusters=4)
assignment = agg.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=assignment, s=60)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

Con el mismo dataset, entrena a un modelo k-means buscando 4 clusters, la predicción de este método es mas similar a la predicción de

```python
from sklearn.cluster import KMeans
import mglearn

# cluster the data into three clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_pred = kmeans.predict(X)

# plot the cluster assignments and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```

Con el mismo dataset, que valor de eps genera exactamente 4 clusters con el algoritmo DBSCAN

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.55)
clusters = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=clusters, s=60)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```
